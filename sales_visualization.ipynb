```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["Sales"].hist(bins=20)         # Sales kis range me sabse zyada hain+  

plt.title("Sales Distribution")
plt.savefig("images/sales_distribution.png")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Sales"].plot(kind="box")      # Agar class me sabke marks 40–80 ke beech hain aur ek student ke 100/100 hain, to wo outlier ho sakta hai.

plt.title("Sales Box Plot")
plt.savefig("images/sales_boxplot.png")
plt.show()
```

```python
df["Sales"].value_counts().head(10)     # Sabse zyada baar kaunsi sales values repeat hui hain.
```

```python
df.groupby("Category")["Sales"].sum()     # Har category ki total sales.
```

```python
df.groupby("Category")["Sales"].sum().plot(kind="bar")

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.groupby("Category")["Sales"].mean()     # Har category ki average sale.
```

```python
df.groupby("Category")["Sales"].mean().plot(kind="bar")

plt.title("Average Sales by Category")
plt.savefig("images/avg_category_sales.png")
plt.xlabel("Category")
plt.ylabel("Average Sales")
plt.show()
```

```python
df.columns
```

```python
df.groupby("Category")["Sales"].agg(["sum", "mean", "max", "min"])
```

```python
df.groupby("Region")["Sales"].sum().plot(kind="bar")

plt.title("Total Sales by Region")
plt.savefig("images/region_sales.png")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.groupby("Segment")["Sales"].sum().plot(kind="bar")

plt.title("Total Sales by Segment")
plt.xlabel("Segment")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.groupby("State")["Sales"].sum().sort_values(ascending=True).head(10)
```

```python
df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Top 10 Sub-Categories by Sales")
plt.savefig("images/top_subcategory_sales.png")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.show()
```
